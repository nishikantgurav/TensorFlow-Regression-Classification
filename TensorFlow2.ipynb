```python
X=df.drop(['price','date'],axis=1).values
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
#preprocessing
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train.shape
```

```python
#model building
import tensorflow as tf
from tensorflow import keras
model=keras.models.Sequential(
    [keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
```

```python
model.fit(x=X_train,y=y_train.values,validation_data=(X_test,y_test.values),epochs=400)
model.history.history
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
predictions=model.predict(X_test)
```

```python
mean_squared_error(y_test,predictions)
```

```python
import numpy as np
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
#keras classification along EDA
df=pd.read_csv('cancer_classification.csv')
df.head()
```

```python
import seaborn as sns
sns.countplot(x='benign_0__mal_1',data=df)
```

```python
df.corr()['benign_0__mal_1'].sort_values()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr())
```

```python
X=df.drop('benign_0__mal_1',axis=1).values
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
model=keras.models.Sequential(
    [keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600)
model.history.history
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop=EarlyStopping(monitor='val_loss',mode='min',patience=25)
```

```python
import tensorflow as tf
from tensorflow import keras
model=keras.models.Sequential(
    [keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[early_stop])
```

```python
model_loss=pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
import tensorflow as tf
from tensorflow import keras
model=keras.models.Sequential(
    [keras.layers.Dense(30,activation='relu'),
     keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[early_stop])
```

```python
model_loss=pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
predictions=model.predict_classes(X_test)
```
